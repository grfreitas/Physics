```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_xy():
    x = random.uniform(0,1)
    y = random.uniform(0,1)
    return x, y
```

```python
N = 100000
inside = 0
R = 1

X = []
Y = []
d = []

for i in range(N):
    x, y = random_xy()
    
    X.append(x)
    Y.append(y)
    
    r = np.sqrt(x**2+y**2)
    d.append(r)
    
    if r <= R:
        inside += 1
        
percent_inside = inside/N
```

```python
colors = np.where(np.array(d)<=R,'red','green')

plt.figure(figsize=(15,6))
plt.scatter(X, Y, c=colors, s=0.1)
plt.axis('equal')
plt.show()
```

```python
percent_inside = inside/N

A_total = 1
A_inside = percent_inside*A_total

A_circ = 4*A_inside

pi = A_circ/R**2

print('Pi Calculado:', pi)
print('Pi Exato    :', np.pi)
print('-'*40)
print('Erro Percentual:', np.abs(pi - np.pi)/np.pi*100,'%')
```
